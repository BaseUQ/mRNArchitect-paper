```python
import polars as pl

from mrnarchitect.sequence import Sequence
```

```python
df = pl.read_csv("optimized-sequences.csv")
df
```

```python
hamming_distance = [
    {
        "index": row["index"],
        "hamming_distance": Sequence.create(
            row["input_sequence"]
        ).hamming_distance(Sequence.create(row["optimized_sequence"])),
        "hamming_ratio": Sequence.create(
            row["input_sequence"]
        ).hamming_distance(Sequence.create(row["optimized_sequence"])) / len(row["input_sequence"]),
    }
    for row in df.rows(named=True)
    if row["optimized_sequence"]
]
output = pl.DataFrame(hamming_distance)
output.write_csv("output-hamming-distance.csv")
output
```
